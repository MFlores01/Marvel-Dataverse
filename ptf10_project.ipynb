```python
box_ofc_p1 = df.query("Phase=='Phase One'")
box_ofc_p2 = df.query("Phase=='Phase Two'")
box_ofc_p3 = df.query("Phase=='Phase Three'")
box_ofc_p4 = df.query("Phase=='Phase Four'")
box_ofc_p5 = df.query("Phase=='Phase Five'")

trace1 = go.Bar(
    x=box_ofc_p1['Film'],
    y=box_ofc_p1['Box office gross Worldwide'],
    name = 'Phase 1',
    marker=dict(color='#FFD700')
)
trace2 = go.Bar(
    x=box_ofc_p2['Film'],
    y=box_ofc_p2['Box office gross Worldwide'],
    name = 'Phase 2',
    marker=dict(color='#9EA0A1')
)
trace3 = go.Bar(
    x=box_ofc_p3['Film'],
    y=box_ofc_p3['Box office gross Worldwide'],
    name = 'Phase 3',
    marker=dict(color='#CD7F32')
)
trace4 = go.Bar(
    x=box_ofc_p4['Film'],
    y=box_ofc_p4['Box office gross Worldwide'],
    name = 'Phase 4',
    marker=dict(color='#0b7314')
)
trace5 = go.Bar(
     x=box_ofc_p5['Film'],
     y=box_ofc_p5['Box office gross Worldwide'],
     name = 'Phase 5',
     marker = dict(color = '#f01707')
)
data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(
    title='Worldwide Box Office by Phases'
)
fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
# @title
# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(df['Budget'], df['Box office gross Worldwide'])

# Create a scatter plot using Plotly Express
scatter_fig = px.scatter(df, x='Budget', y='Box office gross Worldwide', trendline='ols',
                         title='Budget vs Box Office Collection',
                         labels={'Budget': 'Budget (Million USD)', 'Box office gross Worldwide': 'Worldwide Box Office Gross(Million USD)'},
                         color_continuous_scale=px.colors.sequential.Plasma,
                         template='plotly_dark')

# Add Pearson correlation information to the layout
scatter_fig.update_layout(
    annotations=[
        dict(
            x=0.5,
            y=1.1,
            text=f"Pearson correlation (r) = {pearson_corr:.4f}",
            showarrow=False,
            font=dict(color='Pink', size=14),
            xref='paper',
            yref='paper'
        )
    ]
)

# Show the plot
scatter_fig.show()
```

According to the graph, the pearson correlation r value is 0.6924, therefore, it can be inferred there is a strong positive correlation between Budget and Worldwide Box Office Gross. In this case, when Marvel has spent more Budget on production, the results have been positive in Worldwide Box Office Gross. Moreover, every Marvel film that covers up their production budget and has usually brought some massive profits along with it.
